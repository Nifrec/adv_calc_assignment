```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Colormap
```

```python
def plot_coord_lines_const_u_and_v():
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

    # U constant
    ax=axes[0]
    colormap = plt.get_cmap("hsv")
    u_vals = [-3, -1, -0.5, -0.25, 0.25, 0.5, 1, 3]
    circles=[]
    labels=[]
    for i in range(len(u_vals)):
        u = u_vals[i]
        centre = (1/(2*u), 0)
        radius = abs(1/(2*u))

        circle = plt.Circle(centre, radius, fill=False, label=f"$u = {u}$",
        edgecolor=colormap(i/(0.7*len(u_vals))))
        ax.add_artist(circle)
        circles.append(circle)
        labels.append(circle.get_label())

    ax.legend(circles, labels)
    ax.set_xlim([-4, 4])
    ax.set_ylim([-4, 4])
    ax.grid(which="major", alpha=0.6)
    ax.set_title("Coordinate lines for constant u", weight="bold")
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")


    # V constant
    ax=axes[1]
    v_vals = [-3, -1, -0.5, -0.25, 0.25, 0.5, 1, 3]
    circles=[]
    labels=[]
    for i in range(len(v_vals)):
        v = v_vals[i]
        centre = (0, -1/(2*v))
        radius = abs(1/(2*v))

        circle = plt.Circle(centre, radius, fill=False, label=f"$v = {v}$",
        edgecolor=colormap(i/(0.7*len(u_vals))))
        ax.add_artist(circle)
        circles.append(circle)
        labels.append(circle.get_label())

    ax.legend(circles, labels)
    ax.set_xlim([-4, 4])
    ax.set_ylim([-4, 4])
    ax.grid(which="major", alpha=0.6)
    ax.set_title("Coordinate lines for constant v", weight="bold")
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")

    fig.savefig("ass_6_ex_1.pdf")

plot_coord_lines_const_u_and_v()
```
